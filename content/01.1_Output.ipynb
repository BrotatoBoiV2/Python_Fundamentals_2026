# Outputting Information

In Python, you can output data with the built-in `print()` function using the following parameters to display information:

|    Parameter    | Description |
| --------------- | ----------- |
| object(s)       | Takes objects you want to output to the screen. |
| end='end'       | Append a value to the end of the print statement. (Default is '\n') |
| sep='separator' | Insert a value between each object given. |

By the end of this notebook, you should have a basic understanding of how to output data to the screen.

---

# Basic Output

The quickest and most common way to print output is through passing string objects as the `print()` function's argument.
The most basic example would be a "Hello World!" script. You can also output integers and floats as well as multiple objects!

```python
print("Hello World!")
print("Hello", "World!")
print(420)
print(21.69)
```

# Custom Endings

You can select a custom end value for the `print()` function. This defaults to a newline character ('\n') which automatically starts the next print statement on a newline. Changing this value lets you control how the statement actually ends.

```python
print("Hello", end=" ")
print("World!")
```

# Object Separation

Using the `sep` parameter, you can set how the given objects are separated.

```python
print("Hello", "World", sep="#@#")
```
