# User Input

This module teaches how to obtain input from the user and store it into a variable as a string to use later.
By the end, you should have a grasp on obtaining and using user input.

The `input()` function gets user input with these parameters:

| Parameter |            Description             |
| --------- | ---------------------------------- |
|   prompt  | The prompt to display to the user. |

---

```python
username = input("Provide a Username >>> ")

print(username)
```

# Input Conversion

Getting the user input will always return as a string. To solve this issue, you can use the type conversion functions from the Variables module to force the input into the way it needs to be.

***NOTE:*** If you try to type convert without any safety mechanisms in place, there will be an error thrown.

```python
number = int(input("Insert a number >>> ")) # We are assuming the user is playing fair and inserts an integer.

print(number + 3)
```
