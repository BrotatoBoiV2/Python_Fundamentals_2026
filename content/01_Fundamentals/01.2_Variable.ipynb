# Variables

In this module, you will learn how to create and manipulate variables. A variable is sort of a container that
holds an object which can be used and manipulated throughout the script.
A variable is created by typing any name as long as it fits the following criteria:

* Must start with a letter or underscore.
* Cannot start with a number.
* Can only contain Aa-Zz and 0-9 including underscores.
* Name must not exist as a Python keyword or function.
   - Keywords throw an error.
   - Functions break the script - disables function.
* Variables are case-sensitive so be wary.

In Python, variables normally have "snake_case" and should be descriptive of what they are for.

By the end of this module you will understand how to create and manipulate a variable, set the variable to string, integer or float, and reveal which one of those a specific variable is.

---

# Variable Creation

You can create a variable simply with a name, an equals sign and the object/value the variable will hold.
The following are different types of variables that can be made.

Note: The variable name needs to be on the left side of the equals sign.

```python
character_name = "Billy"
current_level = 123
experience_points = 21.30

print(character_name)
print(current_level)
print(experience_points)
```

# Type Conversion

In Python, a type is what the data represents; string, integer, float, etc.
You can convert variables from one type to another as long as it is valid.
For an example, you cannot turn a string "Hello" into an integer since that will throw an error.

```python
coins = 666
coins_string = str(coins)
money = float(coins)

print(coins)
print(coins_string) # No visual difference, but functionally different.
print(money)
```

# Type Discovery

A useful way to find out what the type of a variable is would be to use the built-in `type()` function.
This returns what the variable represents so that you know if the data is processed in the right way.

```python
color = "Yellow"

print(type(color))
```
